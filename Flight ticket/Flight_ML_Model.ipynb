```python
import numpy as np
import pandas as pd
from sklearn import linear_model
```

```python
df = pd.read_csv('Clean_Dataset.csv',index_col=False)

newdf = df.drop(['flight'], axis= ('columns'))
newdf
```

```python
airlin_D = pd.get_dummies(newdf.airline)
airlin_D
```

```python
source_city_D = pd.get_dummies(newdf.source_city)
source_city_D
```

```python
departure_time_D = pd.get_dummies(newdf.departure_time)
departure_time_D
```

```python
stops_D = pd.get_dummies(newdf.stops)
stops_D
```

```python
arrival_time_D = pd.get_dummies(newdf.arrival_time)
arrival_time_D
```

```python
destination_city_D = pd.get_dummies(newdf.destination_city)
destination_city_D
```

```python
class_D = pd.get_dummies(newdf._Class)
class_D
```

```python
merged = pd.concat([df,airlin_D,source_city_D,departure_time_D,stops_D,arrival_time_D,destination_city_D,class_D], axis = ('columns'))
final = merged.drop(['airline','flight','source_city','departure_time','Unnamed: 0','stops','arrival_time','destination_city','_Class'], axis='columns')
#final
```

```python
y = final.price
x = final.drop(['price',],axis=('columns'))
x
```

```python
ml_model = linear_model.LinearRegression()
ml_model.fit(x,y)
accuracy = ml_model.score(x,y)*100
print("ML model Accuracy is",accuracy,'%')
```
